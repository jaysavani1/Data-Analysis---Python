```python
###import all the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os         #for locating the current directory        
import calendar  #for converting month number to month name
from itertools import combinations
from collections import Counter
%matplotlib inline
```

```python
#locate the file in specific folder in PC
files = [file for file in os.listdir('./sales/')]

#create an empty dataframe to merge all the files into a single csv.
all_months_data = pd.DataFrame()

#read and load all the files and continously concatinate the data into a single all_months_data dataframe variable. 
for each_file in files:
    single_each_file = pd.read_csv('./sales/'+each_file)
    all_months_data = pd.concat([all_months_data, single_each_file])
    
#print few raws to understand columns and rows in SALES dataset
display(all_months_data)
```

```python
#check the all null values present in dataset
null_value_check1 = all_months_data.isna()#.any()
display(null_value_check1)
```

```python
#Load main sales.csv file to do analysis
sales  = pd.read_csv ('sales.csv')
sales.head()
```

```python
##create a month column from Order date column

sales['month'] = sales['Order Date'].str[0:2]
sales['month'] = sales['month'].astype('int32')
sales.head()
```

```python
#removing "Or" Error

temp_df = sales[ sales['Order Date'].str[0:2] == 'Or' ]
temp_df
```

## Analysis through problem statements: -

### Question- 1: What is the best month for sale ? and How much money earned in that month ?

### Answer 1 :

```python
#for above question, we need total sales and money values in data. So create total_sales column

sales.info()

#here, we can see all the columns are object or string datatype.(including Quantity Ordered and Price Each columns).
```

```python
#convert the string data columns (all the data in row) into numeric values, so we can create total_sales column.

sales['Quantity Ordered']  = pd.to_numeric(sales['Quantity Ordered'],downcast = 'integer') #convet str to int
sales['Price Each'] = pd.to_numeric(sales['Price Each'], downcast='integer') #convert str to float
```

```python
sales['total_sales'] = sales['Quantity Ordered'] * sales['Price Each']
sales.head()
```

```python
# getting total sales data by each months
sales_by_month = sales.groupby('month').sum()
display(sales_by_month)
```

```python
#creating Months name for x-axis labels in plots.

uniq_months = []
mon = np.array(range(1,13))
for each_month in mon:
    uniq_months.append(calendar.month_name[each_month])
display(uniq_months)
```

```python
#Plot the graph for sales value

plt.bar(uniq_months, sales_by_month['total_sales'])

plt.xlabel('Months')
plt.xticks(rotation = 90)
plt.ylabel('Sales in USD($))')
plt.title('Total sales of 2019')

plt.show()
```

```python
#display chart using seaborn    
sns.barplot(x= uniq_months, y = sales_by_month['total_sales'], data=sales_by_month)
plt.xlabel('Months')
plt.xticks(rotation= 90)
plt.ylabel('Sales in USD($)')
plt.title('Total sales of 2019')
plt.show()
```

From tbe above plots, we can clearly see that, in December, total sales value is the highest.

### Question 2: Which city has the highest number of sale ?

### Answer 2 :

```python
""""
We need a city column for further procedure. And we can create that column using "Purchage Address" column
in sales.csv datafile. To do that,extract the values after second comma(,) present in "Purchage Address" column.
So, it will provide you abbriviation of cities with zipcode/pincode.
"""

sales.head()
```

```python
"""
splitting the Address : there are two methods done using ".apply()" Function.

1. Use direct Lambda function
2. Use a seperate function and then apply to lambda function.

Here I am using 2nd method.

"""
def get_city(address):
    return address.split(',')[1]#the split function splits string and create an array. And [1] gives the value present at index 1. 

def get_state(address):
    return address.split(',')[2].split(' ')[1]

sales['city'] = sales['Purchase Address'].apply(lambda x: f"{get_city(x)}({get_state(x)})") #use lambda function whenever changes happening in cells.

#check the sales dataset for previous changes
display(sales.head())
```

```python
cities= sales['city'].unique()
display(cities)
```

```python
#final answer
sales_by_city = sales.groupby('city').sum()
display(sales_by_city)
```

```python
#plot the graph : Using matplotlib

each_city = [city for city,df in sales.groupby('city')]

plt.bar(each_city, sales_by_city['total_sales'])
plt.xlabel ('City Name')
plt.xticks(rotation =90)
plt.ylabel ('Sales in USD($)')
plt.title ('Sales by cities')
plt.show()
```

```python
#plot the graph using seaborn
sns.set_style('darkgrid')
sns.barplot(x= each_city,
            y= sales_by_city['total_sales'],
            data = sales_by_city,
            color = 'b')
plt.xlabel ('City Name')
plt.xticks(rotation =90)
plt.ylabel ('Sales in USD($)')
plt.title ('Sales by cities')
plt.show()
```

From above graph, we can conclude that San Fransisco spends the highest or sale in san fransisco is the highest among other cities.

### Question3:What time should we display advertisements to maximize likelihood of customer's buying product?

### Answer 3:

```python
#check the data type of an "Order Date" column

sales.info()
```

```python
#convert "Order Date" column's datatype from object to "DateTime" format

sales['Order Date'] = pd.to_datetime(sales['Order Date'])
```

```python
#creat a "hour" and "minute" column by extracting time from "Order Date" column

sales['hour'] = sales['Order Date'].dt.hour
sales['minute'] = sales['Order Date'].dt.minute

#check for newly created columns
display(sales.head())
```

```python
#show the number of sales during each hour segment

number_of_sales_by_hour = sales.groupby('hour').count()
display(number_of_sales_by_hour)
```

```python
#plot the given data by hourly manner : Matplotlib

plt.plot(hours, sales.groupby('hour').count())
plt.xlabel('Time - (Hour)')
plt.ylabel('Number of Orders')
plt.title('Number of items sold at perticular time of day')
plt.grid(True)
plt.show()
```

```python
#plot the given data by hourly manner : Seaborn

sns.set_style('whitegrid')
sns.lineplot(x = hours,
             y = number_of_sales_by_hour['total_sales'],
             data = number_of_sales_by_hour,
             marker = True,
             palette = 'o')
plt.xlabel('Time - (Hour)')
plt.ylabel('Number of Orders')
plt.title('Number of items sold at perticular time of day')
plt.show()
plt.show()
```

So the pick times of day are around at 11 AM and 7 PM.

### Question 4 : What products are most often sold together ?

### Answer 4 :

```python
sales['Product'].unique()
```

```python
#create a dummy/seperate dataframe for same "Order ID" numbers and find out using .duplicated() function.
duplicate_order_df = sales[sales['Order ID'].duplicated(keep=False)]
display(duplicate_order_df)
```

```python
"""
From the above dataframe, we can see that, all the same Order Id has been displayed with same numbers.
we need to add products to same order id numbers in new column called "group". 
"""
duplicate_order_df['group'] = duplicate_order_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
display(duplicate_order_df)
```

```python
duplicate_order_df = duplicate_order_df[['Order ID','group']].drop_duplicates()
display(duplicate_order_df)
```

```python
#to count the sublist from the list, we need to use counter and collection
#https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
count = Counter()

for each_row in duplicate_order_df['group']:
    each_row_list = each_row.split(',')
    count.update(Counter(combinations(each_row_list, 2)))# Here, 2 describes the pair value which is common. If we need most 3 itmes than we have to use 3. 
display(count)
```

```python
#take out the most common or first 10 item pairs bought together.

count.most_common(10)
```

From above we can say that 'Iphone' and 'Lightning Charging Cable' is ithe most often sold together product.

### Question 5: Which product sold the most ? and Why?

### Answer 5:

```python
#filter data according to product
products = sales.groupby('Product')

num_of_product_sale = products.sum()['Quantity Ordered']
num_of_product_sale
```

```python
#create graph for quantity sold of each product
unique_product = [product for product,df in sales.groupby('Product')]
plt.bar(unique_product, num_of_product_sale)
plt.xlabel('Products')
plt.xticks(rotation = 90)
plt.ylabel('Number of Orders')
plt.title('Number of perticular items sold')
plt.grid(True)
plt.show()
```

```python
#find the average price of each product
prices = sales.groupby('Product')['Price Each'].mean()
prices
```

```python
#plotting two plots with different y-axis values and common x-axis values in a single Graph.

fig,ax1 = plt.subplots()

ax1.bar(unique_product, num_of_product_sale)
ax1.set_xlabel('Products')
ax1.set_xticklabels(unique_product,rotation = 90)
ax1.set_ylabel('Number of Orders', color ='b')
ax1.set_title('Number of perticular items sold')

ax2 = ax1.twinx()
ax2.plot(unique_product, prices, color = 'red')
ax2.set_ylabel('Prices', color= 'r')

plt.grid(False)
plt.show()
```

Here we can say that, as the prices are down of products, those are the most sold. Well, it does not correct way to relate but I think it gives the pretty much idea about that.
